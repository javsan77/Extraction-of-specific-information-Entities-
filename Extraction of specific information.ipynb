```python
from transformers import pipeline

 # Use a pipeline to extract entities
nlp = pipeline("ner", model="dbmdz/bert-large-cased-finetuned-conll03-english")

# Text with entities
text = "John lives in New York and works at Google since 2015. At 2020 he already didn't work there.Currently, he works in Amazon but He always remembers that epoc with affection "

 # Extract  entities
entities = nlp(text)
for entity in entities:
  print(f"Texto: {entity['word']}, Etiqueta: {entity['entity']}")
```
